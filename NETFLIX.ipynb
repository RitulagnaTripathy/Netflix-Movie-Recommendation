# NETFLIX MOVIE RECOMMENDATION SYSTEM USING ASSOCIATION RULE BASED LEARNING

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset = pd.read_csv("Netflix Movie Recommendation.csv", header = None)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
transaction = []

for i in range(0,7466):
    transaction.append([str(dataset.values[i,j]) for j in range(0,20)])
```

**Apriori training on dataset**

```python
from apyori import apriori
movie_intelligence = apriori(transactions = transaction, min_support = 0.002, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(movie_intelligence) #raw data
```

```python
#Organizing result

def inspect(results):

    movie1         = [tuple(result[2][0][0])[0] for result in results]
    movie2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]

    return list(zip(movie1, movie2, supports))

DataFrame_intelligence = pd.DataFrame(inspect(results), columns = ['movie1', 'movie2', 'Support'])
```
